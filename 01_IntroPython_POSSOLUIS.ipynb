```python
# Codigo del ejercicio
import pandas as pd

#Lector del archivo
df = pd.read_csv("/content/titanic.csv")

# Estadísticas generales con "describe()"
print("Estadísticas descriptivas:")
print(df[['Age','Fare','SibSp','Parch']].describe())

# Porcentaje de pasajeros que sobrevivieron
survived_percentage = df['Survived'].mean() * 100
print(f"\n Porcentaje de sobrevivientes: {survived_percentage:.2f}%")

# Porcentaje de pasajeros según su clase
class_percentage = df['Pclass'].value_counts(normalize=True) * 100
print("\n Porcentaje de pasajeros por clase:")
print(class_percentage)

# Número de hombres y mujeres que viajaron
gender_counts = df['Sex'].value_counts()
print("\n Cantidad de hombres y mujeres:")
print(gender_counts)

#  Porcentaje de hombres y mujeres
gender_percentage = df['Sex'].value_counts(normalize=True) * 100
print("\n Porcentaje de hombres y mujeres:")
print(gender_percentage)


```
